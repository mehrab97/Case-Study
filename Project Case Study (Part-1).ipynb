## Number of Fundings
Given File 'startup_funding.csv'
Problem Statement :
Check the trend of investments over the years. To check the trend, find -
Total number of fundings done in each year.
Plot a line graph between year and number of fundings. Take year on x-axis and number of fundings on y-axis.
Print year-wise total number of fundings also. Print years in ascending order.
Note :
There is some error in the 'Date' feature. Make sure to handle that.
Output Format :
year1 TotalFunding1
year2 TotalFunding2
. . . 

```python
import matplotlib.pyplot as plt
import csv
file_obj=open('/Users/shabeggill/Downloads/startup_funding.csv')
file_data=csv.reader(file_obj,skipinitialspace=True)
file_list=list(file_data)
count1=count2=count3=0
for i in range(1,len(file_list)):
    if "2015" in file_list[i][1]: #Have considered even those columns where funding amount is missing
        count1+=1                 #Need to get the number of fundings done in a year and not amount
    elif "2016" in file_list[i][1]:
        count2+=1
    elif "2017" in file_list[i][1]:
        count3+=1
total_number_fundings=[count1,count2,count3]
year=[2015,2016,2017]
plt.plot(year,total_number_fundings)
plt.xlabel('Year')
plt.ylabel('Number of Fundings')
plt.axis([2015,2017,400,1000])  # first two entries signify min & max x values and last two entries signify min and max y 
plt.xticks(rotation=45)
plt.show()
    
print(year[0],total_number_fundings[0])
print(year[1],total_number_fundings[1])
print(year[2],total_number_fundings[2])
```

## Investment Type
Given File 'startup_funding.csv'
Problem Statement :
There are 4 different type of investments. Find out percentage of amount funded for each investment type.
Plot a pie chart to visualise.
Print the investment type and percentage of amount funded with 2 decimal places after rounding off.
Note :
Correct spelling of investment types are - "Private Equity", "Seed Funding", "Debt Funding", and "Crowd Funding". Keep an eye for any spelling mistake. You can find this by printing unique values from this column.
Print the investment type in descending order with respect to the percentage of the amount funded.
Output Format :
investmentType1 percent1
investmentType2 percent2
investmentType3 percent3
. . . 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data=pd.read_csv('/Users/shabeggill/Downloads/startup_funding.csv')

data['InvestmentType'].replace('PrivateEquity','Private Equity',inplace=True)
data['InvestmentType'].replace('SeedFunding','Seed Funding',inplace=True)
data['InvestmentType'].replace('Crowd funding','Crowd Funding',inplace=True)
data['InvestmentType'].unique()
data.AmountInUSD.fillna('0',inplace=True)     
data['AmountInUSD']=data['AmountInUSD'].str.replace(",","")
data['AmountInUSD']=data['AmountInUSD'].astype(int)
grouped=data.groupby('InvestmentType').AmountInUSD.sum()
grouped=grouped.sort_values()[::-1]
funding=[]
investment=['Private Equity','Seed Funding','Debt Funding','Crowd Funding']
for i in grouped:
    funding.append(i) #--> appending funding amount in the array investment wise
np_funding=np.array(funding) #--> creating a numpy funding array which can be used for total funding sum purpose
explodes=[0,0.2,0.5,0.9]
plt.pie(np_funding,explode=explodes,labels=investment,autopct='%.2f%%')
plt.axis('equal')
plt.show()
print(grouped) #--> gives Investment type along with the funding amount
total=np.sum(funding) #--> total funding amount 
for i in range(len(np_funding)):
    print(investment[i],format(round(100*np_funding[i]/total,2),'.2f')) #--> printing the investment type and coressponding funding %
```

industry1 percent1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data=pd.read_csv('/Users/shabeggill/Downloads/startup_funding.csv')
data.AmountInUSD.fillna('0',inplace=True)     
data['AmountInUSD']=data['AmountInUSD'].str.replace(",","") #--> removing commas from funding amounts
data['AmountInUSD']=data['AmountInUSD'].astype(int) #--> converting the vlues to int
#data['IndustryVertical'].nunique() # to check how Ecommerce is incorrectly used in the dataset and replacing them as below
data['IndustryVertical'].replace('ecommerce','Ecommerce',inplace=True)
data['IndustryVertical'].replace('ECommerce','Ecommerce',inplace=True)
data['IndustryVertical'].replace('eCommerce','Ecommerce',inplace=True)
grouped=data.groupby('IndustryVertical').AmountInUSD.sum() #--> taking sum of funding amount industry vertical wise
grouped=grouped.sort_values()[::-1] # sorting the values in descending order
#print(grouped) --> will five industry vertical along with funding in descending order
funding=[]
industry_vertical=['Ecommerce','Consumer Internet','Technology','Online Marketplace','E-Commerce & M-Commerce platform']
for i in grouped[:5]:
    funding.append(i) # appending top 5 funding amounts
np_funding=np.array(funding)
total=np.sum(np_funding) # taking total of funding amount of the five inudstry verticals
for i in range(len(np_funding)):
    print(industry_vertical[i],format(round(100*np_funding[i]/total,2),'.2f'))
```
